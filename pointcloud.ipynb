```python
import cv2
import datetime
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
from pandas.io.common import file_exists
file_path="../content/drive/MyDrive/Dataset/sequences/00/"
velodyne_file=os.listdir(file_path+'velodyne')
pointcloud=np.fromfile(file_path+'velodyne/'+velodyne_file[0] , dtype=np.float32)
```

```python
pointcloud=pointcloud.reshape((-1,4))
pointcloud.shape
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111 , projection='3d')
zs=pointcloud[:,0]
ys=pointcloud[:,1]
xs=pointcloud[:,2]
ax.scatter(zs,ys,xs , s=0.01)
```

```python
pointcloud.dtype
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111 , projection='3d')
zs=pointcloud[:,0]
ys=pointcloud[:,1]
xs=pointcloud[:,2]
ax.grid('false')
ax.axis('off')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=40, azim=100)
ax.scatter(zs,ys,xs , s=0.0001)
```
